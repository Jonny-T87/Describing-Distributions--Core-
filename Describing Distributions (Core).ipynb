```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('Data/data (1).csv')
df.head()
```

```python
features = df[['Doc_visits', 'VitD_levels', 'TotalCharge']]
features.head()
```

- Doc_visits histogram 

```python
##making mean and median and std for doc visits with hisplot
col = 'Doc_visits'
fea = df[col]
doc_mean = fea.mean()
doc_median = fea.median()
doc_std = fea.std()
plus_one_std = doc_mean + doc_std
minus_one_std = doc_mean - doc_std

fig,ax = plt.subplots(figsize=(10,6))
ax = sns.histplot(data=features, x=fea, stat='probability', kde=True)
ax.axvline(doc_mean, color='red', label=f'Mean = {doc_mean:,.2f}')
ax.axvline(doc_median, ls='--', color='green', label=f'Median = {doc_median:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color='yellow', zorder=0)
ax.set_title(f'{col}')
ax.legend();
```

1. Is it Discrete or Continuous?
   - Based on the plot this is discrete and also because the distributions can only take on certain values. Not between any value.
   
2. Does it have a skew? If so, which direction (+/-)
    - The data is not skewed because the mean and median are basically the same.
    
3. What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

  - This displays a Leptokurtic because values cluster in the peak of the distribution.
